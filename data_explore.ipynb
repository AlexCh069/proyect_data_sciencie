```python
# 1. Detectamos todos los features tipo 'P2XX' o 'P3XX' 

# Lista de ejemplo
columnas = df.columns

# Filtrar columnas que comienzan con 'P2', 'P3' Y 'P4'
filtrados = [dato for dato in columnas if dato.startswith(('P2', 'P3','P4'))]

print("columnas filtrados:", filtrados)
```

```python
# Ya que el formato de estos datos es una mezcla de 'str' e 'int', necesitamos que solo presente
# valores tipo int, ya que es el dato adecuado para estos features


# Df para pruebas
p1 = df[filtrados[0:5]]

data_values(p1)
```

```python
import pandas as pd

def mod_col(data:pd.DataFrame, name_col):

    # Si es int64 no modificamos nada
    if data[name_col].dtype == 'int64':
        return data[name_col]

    # Si es string y contiene un numero, lo pasamos a int

    valores = list(data[name_col].values)
    valores_modificados = []

    for val in valores:
        if type(val) == 'int64':
            valores_modificados.append(val)

        else:
            # Si es un string que contiene un numero, lo pasamos a int
            try:
                valores_modificados.append(int(val))
            except ValueError:
                # Si no es un numero, lo dejamos como esta
                valores_modificados.append(None)
    
    return pd.Series(valores_modificados)
    
```

```python
for col in p1.columns:
    mod_col(p1, col)
    p1[col] = mod_col(p1, col)
```
